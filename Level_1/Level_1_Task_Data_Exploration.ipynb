# Level 1 - Task 1: Data Exploration and Preprocessing

## Objective
The objective of this task is to explore the restaurant dataset, understand its structure, identify missing values, check data types, and analyze the distribution of the target variable (Aggregate Rating).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load the Dataset

```python
import pandas as pd

df = pd.read_csv("../Dataset/Dataset.csv")

df.head()
```

## Dataset Shape

```python
# Display the number of rows and columns
print("Shape of Dataset:", df.shape)
```

## Dataset Information

```python
# Display dataset information
df.info()
```

## Data Types

```python
# Data types
print(df.dtypes)
```

## Missing Values

```python
# Missing values
print("Missing Values:")
print(df.isnull().sum())
```

## Checking Missing Values

The dataset was checked for missing values in all columns using the `isnull().sum()` function.
This helps identify columns that contain null values and need preprocessing before analysis.

```python
# Check missing values
df.isnull().sum()
```

### Observation

- Most columns do not contain any missing values.
- Only the **Cuisines** column contains **9 missing values**.
- Since only 9 out of 9551 records have missing values, these rows can be safely removed without significantly affecting the dataset.

## Handling Missing Values

Since only the **Cuisines** column contained **9 missing values**, these records were removed using the `dropna()` function. The dataset was then checked again to ensure that no missing values remained.

```python
df = df.dropna()

print("Dataset Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())
```

## Observation

- The dataset contained **9 missing values**, all in the **Cuisines** column.
- Since these missing records represented less than **0.1%** of the dataset (9 out of 9551 rows), they were removed using the `dropna()` function.
- After preprocessing, the dataset contains **9542 rows**, **21 columns**, and **no missing values**.

## Distribution of Aggregate Rating

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Aggregate rating"], bins=10)

plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")

plt.show()
```

## Observation

- Most restaurants have ratings between 3.0 and 4.0.
- A large number of restaurants have a rating of 0, indicating unrated restaurants.
- The distribution is not perfectly balanced, showing a slight class imbalance.

## Observation

- The dataset originally contained 9551 rows and 21 columns.
- There were 9 missing values in the "Cuisines" column.
- Missing values were handled by removing those rows using `dropna()`.
- After preprocessing, the dataset contains 9542 rows and 21 columns.
- No missing values remain in the dataset.

# Level 1 - Task 3: Geospatial Analysis

## Objective

The objective of this task is to analyze the geographical distribution of restaurants using their latitude and longitude coordinates. It also examines how restaurants are distributed across different cities and countries and explores whether there is any relationship between a restaurant's location and its aggregate rating.

## Restaurant Locations on the Map

This section visualizes the geographical locations of restaurants using their latitude and longitude coordinates.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(
    df["Longitude"],
    df["Latitude"],
    alpha=0.5
)

plt.title("Restaurant Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```

## Observation

- The scatter plot shows the geographical distribution of restaurants based on their latitude and longitude coordinates.
- Restaurants are spread across multiple regions, indicating that the dataset includes restaurants from different countries and cities.
- Several dense clusters of points are visible, suggesting that a large number of restaurants are concentrated in specific geographic locations.
- Some locations contain only a few restaurants, resulting in isolated data points on the map.
- Overall, the dataset has a wide geographical coverage, with restaurants distributed across different parts of the world.

## Correlation Between Restaurant Location and Aggregate Rating

This section examines whether there is any relationship between the geographical location of restaurants and their aggregate ratings.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Longitude"],
    df["Aggregate rating"],
    alpha=0.5
)

plt.title("Longitude vs Aggregate Rating")
plt.xlabel("Longitude")
plt.ylabel("Aggregate Rating")

plt.show()
```

## Observation

- The scatter plot does not show any strong relationship between **longitude** and **aggregate rating**.
- Restaurants with both high and low ratings are distributed across different longitude values.
- Most restaurants have aggregate ratings between **3.0 and 4.5**, regardless of their longitude.
- A few restaurants have an aggregate rating of **0**, representing unrated restaurants, which are also spread across different locations.
- Overall, longitude does not appear to have a significant influence on restaurant ratings.

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Latitude"],
    df["Aggregate rating"],
    alpha=0.5
)

plt.title("Latitude vs Aggregate Rating")
plt.xlabel("Latitude")
plt.ylabel("Aggregate Rating")

plt.show()
```

## Observation
