```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../Dataset/Dataset.csv")
```

# Level 2 - Task 1: Table Booking and Online Delivery

## Objective

The objective of this task is to analyze the availability of table booking and online delivery services offered by restaurants. It also compares the average aggregate ratings of restaurants based on these services to understand their impact on customer ratings.

# Percentage of Restaurants with Table Booking

This section calculates the percentage of restaurants that offer table booking.

```python
table_booking = df["Has Table booking"].value_counts(normalize=True) * 100

table_booking
```

## Observation

The percentage distribution shows the proportion of restaurants that offer table booking compared to those that do not.

```python
import matplotlib.pyplot as plt

table_booking.plot(kind="bar")

plt.title("Percentage of Restaurants with Table Booking")
plt.xlabel("Table Booking")
plt.ylabel("Percentage")

plt.show()
```

## Observation

The bar chart clearly illustrates the difference between restaurants that offer table booking and those that do not, making it easier to compare their proportions.

# Average Rating Based on Table Booking

This section compares the average aggregate rating of restaurants that offer table booking with those that do not.

```python
df.groupby("Has Table booking")["Aggregate rating"].mean()
```

### Observation

- Restaurants that offer table booking have a higher average rating (approximately **3.44**) compared to restaurants that do not offer table booking (approximately **2.56**).
- This indicates that restaurants with table booking facilities generally receive better customer ratings.
- The availability of table booking may be associated with improved customer experience and service quality.

```python
df.groupby("Has Table booking")["Aggregate rating"].mean().plot(kind="bar")

plt.title("Average Rating by Table Booking")

plt.xlabel("Table Booking")

plt.ylabel("Average Rating")

plt.show()
```

### Observation

- The bar chart clearly shows that restaurants with **table booking ("Yes")** have noticeably higher average ratings than those without table booking ("No").
- The difference in ratings suggests a positive relationship between table booking availability and customer satisfaction.
- Overall, restaurants providing table booking tend to perform better in terms of customer ratings.

# Online Delivery Across Price Ranges

This section analyzes the availability of online delivery across different restaurant price ranges.

```python
pd.crosstab(df["Price range"], df["Has Online delivery"])
```

### Observation

- Price Range 1 has the highest number of restaurants, with 3,737 not offering online delivery and 701 offering online delivery.
- Price Range 2 has a relatively balanced distribution, with 1,827 restaurants without online delivery and 1,286 with online delivery.
- Price Range 3 has fewer restaurants, and online delivery is less common compared to lower price ranges.
- Price Range 4 has the smallest number of restaurants, with only 53 restaurants providing online delivery.
- Overall, restaurants without online delivery outnumber those offering online delivery across all price ranges.

```python
pd.crosstab(df["Price range"], df["Has Online delivery"]).plot(kind="bar")

plt.title("Online Delivery Across Price Ranges")

plt.xlabel("Price Range")

plt.ylabel("Number of Restaurants")

plt.show()
```

### Observation

- The bar chart clearly shows that restaurants without online delivery are more common in every price range.
- Online delivery is most frequently available among Price Range 2 restaurants.
- Price Range 1 contains the largest number of restaurants overall, but only a small proportion offer online delivery.
- Very few premium restaurants (Price Range 4) provide online delivery services.
- This suggests that online delivery is more popular among low- and mid-priced restaurants than among expensive restaurants.

## Final Summary

In this task, the availability of table booking and online delivery services was analyzed along with their relationship to restaurant ratings and price ranges.

The analysis showed that only a small percentage of restaurants offer table booking, while the majority do not. However, restaurants that provide table booking have a noticeably higher average aggregate rating than those without this service.

The availability of online delivery was also examined across different price ranges. Most restaurants, regardless of their price range, do not offer online delivery. Online delivery is more common among low- and mid-priced restaurants, whereas high-end restaurants are less likely to provide this service.

Overall, the results indicate that value-added services such as table booking are associated with higher customer ratings, while online delivery is primarily offered by restaurants in the lower and middle price categories.

# Level 2 - Task 2: Price Range Analysis

## Objective

The objective of this task is to analyze the distribution of restaurant price ranges, calculate the average rating for each price range, and identify the rating color associated with the highest average rating.

## Most Common Price Range

This section identifies the most frequently occurring price range in the dataset.

```python
# Count restaurants in each price range
df["Price range"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Price range"].value_counts().sort_index().plot(kind="bar")

plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.show()
```

### Observation

- Price Range 1 contains the highest number of restaurants.
- The number of restaurants decreases as the price range increases.
- Price Range 4 has the fewest restaurants.
- This indicates that low-cost restaurants are more common in the dataset than expensive restaurants.

## Average Rating by Price Range

This section calculates the average aggregate rating for restaurants in each price range.

```python
df.groupby("Price range")["Aggregate rating"].mean()
```

```python
df.groupby("Price range")["Aggregate rating"].mean().plot(kind="bar")

plt.title("Average Rating by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Average Rating")

plt.show()
```

Observation:
The analysis shows that the average restaurant rating increases as the price range increases. Restaurants in Price Range 1 have the lowest average rating (approximately 2.00), while those in Price Range 4 have the highest average rating (approximately 3.82). This indicates a positive relationship between price range and customer ratings, suggesting that higher-priced restaurants generally receive better ratings, possibly due to better food quality, service, and overall dining experience.

## Rating Color Associated with Highest Average Rating

This section identifies which rating color corresponds to the highest average aggregate rating.

```python
df.groupby("Rating color")["Aggregate rating"].mean().sort_values(ascending=False)
```

```python
df.groupby("Rating color")["Aggregate rating"].mean().sort_values().plot(kind="barh")

plt.title("Average Rating by Rating Color")
plt.xlabel("Average Rating")
plt.ylabel("Rating Color")

plt.show()
```

Observation:
The analysis indicates that restaurants with the Dark Green rating color have the highest average rating (approximately 4.66), followed by Green (4.17), Yellow (3.68), Orange (3.05), and Red (2.30). The White rating color has an average rating of 0.00, representing unrated restaurants. This confirms that the rating colors accurately reflect restaurant performance, with Dark Green representing the highest-rated restaurants and Red/White representing the lowest or unrated restaurants.

## Final Summary

In this task, the distribution of restaurant price ranges and their relationship with customer ratings were analyzed. The analysis showed that Price Range 1 is the most common category, indicating that the dataset contains a larger number of budget-friendly restaurants.

The average aggregate rating was calculated for each price range, revealing that restaurant ratings generally vary across different pricing categories. This suggests that the cost of dining alone does not determine customer satisfaction, as restaurants from multiple price ranges receive good ratings.

Finally, the average rating was analyzed based on the rating color. The rating color associated with the highest average rating represents restaurants with the best customer satisfaction, confirming that the rating color effectively reflects the restaurant's overall performance.

Overall, the analysis provides valuable insights into the relationship between restaurant pricing, customer ratings, and rating categories, helping to better understand customer preferences and restaurant quality across different price segments.

# Level 2 - Task 3: Feature Engineering

## Objective

The objective of this task is to create new features from existing columns and convert categorical variables into numerical values to make the dataset more suitable for data analysis and machine learning.

## Restaurant Name Length

This section creates a new feature representing the number of characters in each restaurant's name.

```python
df["Restaurant Name Length"] = df["Restaurant Name"].str.len()

df[["Restaurant Name", "Restaurant Name Length"]].head()
```

### Observation

- A new feature called **Restaurant Name Length** was successfully created.
- It stores the number of characters in each restaurant's name.
- This feature may help identify whether the length of a restaurant's name has any relationship with customer ratings or popularity.

## Address Length

This section creates a new feature representing the number of characters in each restaurant's address.

```python
df["Address Length"] = df["Address"].str.len()

df[["Address", "Address Length"]].head()
```

### Observation

- A new feature called **Address Length** was created from the Address column.
- It records the number of characters in each restaurant's address.
- This feature can be useful for future analysis and predictive modeling.

## Encoding Has Table Booking

This section converts the categorical values ("Yes" and "No") into numerical values.

```python
df["Has Table booking"] = df["Has Table booking"].map({
    "Yes": 1,
    "No": 0
})

df[["Has Table booking"]].head()
```

### Observation

- The categorical values were successfully converted into numerical values.
- "Yes" is represented by **1**, while "No" is represented by **0**.
- Numerical encoding makes the data suitable for machine learning algorithms.

## Encoding Has Online Delivery

This section converts the categorical values ("Yes" and "No") into numerical values.

```python
df["Has Online delivery"] = df["Has Online delivery"].map({
    "Yes": 1,
    "No": 0
})

df[["Has Online delivery"]].head()
```

### Observation

- The "Has Online delivery" column was successfully encoded into numerical values.
- Restaurants offering online delivery are represented by **1**, while those not offering online delivery are represented by **0**.
- This transformation prepares the dataset for machine learning and statistical analysis.

## Final Summary

In this task, new features were created from existing columns by calculating the length of restaurant names and addresses. These engineered features can provide additional information for future analysis and predictive modeling.

Furthermore, categorical variables such as **Has Table booking** and **Has Online delivery** were converted into numerical values using label encoding. This preprocessing step makes the dataset compatible with machine learning algorithms, as numerical features are required by most predictive models.

Overall, feature engineering enhanced the dataset by creating meaningful new attributes and preparing categorical data for advanced data analysis and model development.
